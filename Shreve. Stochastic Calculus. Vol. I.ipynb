# <font color=7E87FC>1. The Binominal No-Artitrage Pricing Model</font>
## <font color=7E87FC>1.1 One-Period Binominal Model</font>
### В простейшей однопериодной модели есть <font color=7E87FC>time zero</font> в начале и <font color=7E87FC>time one</font> в конце. В $t=0$ есть цена акции $S_0 > 0$. В $t=1$ цена акции будет либо $S_1(H)$ (heads) или $S_1(T)$ (tail) – предположим, что как будто подбрасывание монетки опр. цену акции в $t=1$, причём "probability is not assumed to be fair ($\neq 0.5$))". Пусть вероятность heads будет $p$, а tail – $q=1-p$.
<img src='https://sun9-24.userapi.com/impg/PmxCWMWnDGh5EAOd93Sx7YCVVXsYJtuc4gv8WQ/Q49lwEdaE7g.jpg?size=333x250&quality=96&sign=0099371d82f27202dae9d33f2203855e&type=album' width=250 align='right' vspace="15" hspace="20"></img>
### Цена на акцию будет рандомной, т.к. мы не знаем, какой исход будет в $t=1$. Также вводим две положительные величины: $u=\frac{S_1(H)}{S_0} > d=\frac{S_1(T)}{S_0}$ – up- and down-факторы. Также мы вводим <font color=7E87FC>intereset rate</font> $r$. Один доллар, вложенный на денежный рынок в $t=0$, даст $(1+r)$ в момент $t=1$. Также можно занимать на рынке, и в $t=1$ будет долг $(1+r)$. Для дальнейших рассуждений предположим, что $r>-1$.
### Ключевая фишка <font color=7E87FC>efficient market</font> заключается в том, что если есть возможность сделать деньги из воздуха, также должен сущ. риск потерь. Иначе был бы <font color=7E87FC>arbitrage</font> – такой рынок, на котором есть нулевая вероятность риска и положительная вероятность сделать деньги. 
### Важное неравенство: $0<d<1+r<u$ – <font color=7E87FC>non-arbitrage</font>.
### <font color=7E87FC>European call option</font> – право, но не обязательство купить одну акцию в момент $t=1$ по <font color=7E87FC>strike price</font> $K$. Также $S_1(T)<K<S_1(H)$. То есть если у нас выпадают tail, то опцион бесполезен. Если же heads, то исполняем опцион и зарабатываем $S_1(H)-K$. 
### В $t=1$ опцион стоит $(S_1-K)^+$, где запись $(\ldots)^+$ обозначает, что мы берём максимум от выражения в скобках. Фундаментальный вопрос – сколько стоит на самом деле опцион в $t=0$ до того, как мы подбросим монетку в $t=1$.
### <font color=7E87FC>Пример 1.1.1</font>. Допущения: актив делим (с большими суммами и лотами можно повторить в жизни), процентная ставка = const (для больших учреждений), цена продажи и покупки одинаковая (нет bid-ask spread – fake in real life), цена акции в $t=1$ принимает одно из двух значений (такие же, кстати, есть у Блэка-Шоулза кроме четвёртого – там Geometric Brownian Motion). Европейский опцион-колл: $S(0)=4, u=2, d=\frac{1}{2}, r=\frac{1}{4}, S_1(H)=8, S_1(T)=2, K=5$. Также у нас есть изначальный капитал $X_0 = 1.2$ и мы покупаем $\Delta_0 = \frac{1}{2}$ долю от акции (идеально делимая) в $t=0$. В момент $t=0$ мы покупаем половину акции на изначальный запас капитала и также берём в долг на денежном рынке, то есть наша <font color=7E87FC>cash position</font> равна $X_0 - \Delta_0S_0 = -0.8$ (т.е. долг в 0.8 на денежном рынке). В момент $t=1$ наша cash position будет $(1+r)(X_0-\Delta_0S_0)=-1$. В момент $t=1$ наша акция будет стоить $\frac{1}{2}S_1(H) = 4$ или $\frac{1}{2}S_1(T)=1$. 
### В случае, если выпадет heads, то наш портфель будет $X_1(H)=\frac{1}{2}S_1(H)+(1+r)(X_0-\Delta_0S_0)=3$.
### Если же tail, то $X_1(T) = \frac{1}{2}S_1(T)+(1+r)(X_0-\Delta_0S_0)=0$.
### А если у нас будет опцион на акцию, то будет $(S_1(H)-5)^+=3$ или $(S_1(T)-5)^+=0$. Мы воспроизвели опцион, торгуя на фондовом и денежном рынках.
### Изначальный капитал в $1.2$, которое помогает нам воспроизвести портфель – <font color=7E87FC>no-arbitrage price of the option</font> в $t=0$. Если бы трейдер продал опцион за, например, $1.21$, то он мог быть инвестировать избыточные $0.01$ в денежный рынок и использовать оставшиеся $1.2$, чтобы реплицировать опцион. В момент $t=1$ продавец сможет pay off опцион, несмотря на исход броска монетки, и по-прежнему иметь $0.0125$ от вложений в денежный рынок. Это арбитраж, т.к. продавцу опциона не нужно денег изначально, и без риска потерь он сможет залутать $0.0125$ в $t=1$. С другой стороны, если бы можно было купить опцион по цене ниже, например, $1.19$, тогда он купил бы опцион и реплицировать стратегию выше так же.
### Получается, что в данном случае есть арбитраж, если цена опциона в $t=0$ отличается от $1.2$. Если не отличается, то тогда нет арбитража.
### В общем виде обозначим <font color=7E87FC>derivative security</font> (ПФИ) – ценная бумага, которая платит нам $V_1(H)$ в $t=1$, если монетка выдаёт heads и также $V_1(T)$ при tail. Такж есть <font color=7E87FC>European put option</font>, который платит $(K-S_1)^+$ в $t=1$, где $K=const$; и <font color=7E87FC>forward contract</font>, цена которого в $t=1$ будет $S_1-K$.
### Чтобы опр. цену $V_0$ в $t=0$ для дериватива, мы реплицируем его как вримере 1.1.1. Допустим, у нас есть капитал $X_0$ и мы покупаем $\Delta_0$ долей акции в $t=0$, то есть наша cash position $X_0-\Delta_0S_0$. Стоимость нашего портфеля акций и средств на денежном рынке будет $X_1= \Delta_0S_1 + (1+r)(X_0-\Delta_0S_0) = (1+r)X_0 + \Delta_0(S_1-(1+r)S_0)$ (с $(1=r)$ мы либо одалживаем, либо берём в долг, в завис. от cash position).

### Мы хотим выбрать $X_0, \Delta_0$ так, чтобы $X_1(H)=V_1(H)$ и $X_1(T)=V_1(T)$ ($V_1$ – то, что нам дериватив pays off, в зависимости от исхода броска монетки). Нам известны вероятности этих двух событий, но мы не знаем, какой будет исход. Репликация дериватива будет:
### $X_0 + \Delta_0(\frac{1}{1+r}S_1(H)-S_0) =\frac{1}{1+r}V_1(H)$
### $X_0 + \Delta_0(\frac{1}{1+r}S_1(T)-S_0) =\frac{1}{1+r}V_1(T)$
### Умножим первое на $\tilde{p}$, второе – на $\tilde{q}=1-\tilde{p}$ и сложим вместе:
### $X_0+\Delta_0(\frac{1}{1+r}[\tilde{p}S_1(H)+\tilde{q}S_1(T)]-S_0) = \frac{1}{1+r}[\tilde{p}V_1(H)+\tilde{q}V_1(T)]$
### Если мы выберем $\tilde{p}$ такое, что $S_0 = \frac{1}{1+r}[\tilde{p}S_1(H)+\tilde{q}S_1(T)]$, тогда слева при умножении слагаемого на $\Delta_0$ получается ноль, и у нас будет простая формула:
### $X_0 = \frac{1}{1+r}[\tilde{p}V_1(H)+\tilde{q}V_1(T)]$
### Можем напрямую найти $\tilde{p}$ в форме $S_0 = \frac{1}{1+r}[\tilde{p}uS_0+(1-\tilde{p})dS_0] = \frac{S_0}{1+r}[(u-d)\tilde{p}+d]$
### $\tilde{p} = \frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$
### Можем решить для $\Delta_0$ и получить <font color=7E87FC>delta-hedging formula</font>: $\Delta_0 = \frac{V_1(H)-V_1(T)}{S_1(H)-S_1(T)}$
### Если трейдер начинает с капиталом $X_0$ и покупает $\Delta_0$ долей акций, то если на монетке выпадет heads, то у трейдера будет портфель $V_1(H)$, а если tail, то $V_1(T)$. Трейдер захеджировал короткую позицию по деривативу. Дериватив, который платит $V_1$ должен будет в $t=0$ стоить $V_0 = \frac{1}{1+r}[\tilde{p}V_1(H)+\tilde{q}V_1(T)]$.
### В биноминальных моделях для деривативов важны сами величины up- и down-величин, а не вероятности, которые им приписаны. В моделях с непрерывным временем цена дериватива зависит от волатильности акции, но не от средних темпов роста.

## <font color=7E87FC>1.2 Multiperiod Binominal Model</font>
### Тут эксперимент с монеткой постоянно повторяется и также двигается up или down. На денежном рынке процентная ставка $r$ постоянная. Также допущение – no-arbitrage condition.
<img src='https://sun9-80.userapi.com/impg/6NJ0FEYjAqyWDmCFL0sNluKF6mJx7jfSZf3FBQ/45VU2wa0aoc.jpg?size=787x451&quality=96&sign=61d6c7d4b8eab8196cf74dbd2515b2fb&type=album' width=280 align='right' vspace="10" hspace="20"></img>
### Начальная цена акции $S_0 > 0$. В $t=1$ имеем $S_1(H)=uS_0$, либо $S_1(T)=dS_0$. На второй бросок монетки цена будет одна из $S_2(HH)=uS_1(H)=u^2S_0, S_2(HT)=duS_0$, $S_2(TH)=udS_0, S_2(TT)=d^2S_0$. На третий бросок будет уже 8 исходов, и не все будут разными. 
### <font color=7E87FC>Пример 1.2.1</font>. Трёхпериодная модель с $S_0=4, u=2,d=\frac{1}{2}$:
<img src='https://sun9-86.userapi.com/impg/KRfl_8oDwZxdsUxYk0U_JGnqLa49iIydjwKDkQ/2osGQ8MRHjA.jpg?size=826x442&quality=96&sign=6e553f349493cc6fc061678eab49c79f&type=album' width=350 align='right' vspace="10" hspace="20"></img>
### Рассмотрим европейский колл-опцион со страйком $K$ в момент $t=2$. Расширим анализ до арбитражного европейского дериватива с датой экспирации $N≥2$. В дату экспирации payoff колл-опциона со страйком $K$ и эксирпацией в момент 2, будет $V_2 = (S_2-K)^+$, где $V_2,S_2$ зависят от первого и второго бросков монеты. Нужно опр. неарбитражную цену опциона в $t=0$. Допустим, трейдер продаёт опцион в $t=0$ по $V_0$ долларов (ещё предстоит определить). Далее трейдер покупает $\Delta_0$ долей акции, инвестируя/занимая $V_0-\Delta_0S_0$ долларов на денежном рынке, чтобы профинансировать это действие. 
### В $t=1$ портфель трейдера (исключая короткие позиции в опционе) будет $X_1 = \Delta_0S_1+(1+r)(V_0-\Delta_0S_0)$. $S_1$ и $X_1$ зависят от исхода первого броска монеты. Тогда будет два уравнения:
### $X_1(H) = \Delta_0S_1(H)+(1+r)(V_0-\Delta_0S_0)$
### $X_1(T) = \Delta_)S_1(T)+(1+r)(V_0-\Delta_0S_0)$
### После первого броска трейдер будет иметь портфель $X_1$ и может перестроить его hedge. Допустим, теперь он хочет держать $\Delta_1$ долей акции, где эта величина зависит от первого броска, т.к.трейдер знает результаты этого броска в $t=0$. Остаток портфеля $X_1-\Delta_1S-1$ опять инвестируется в денежный рынок. В следующий период времени:
### $V_2 = \Delta_1S_2 + (1+r)(X_1-\Delta_1S_1)$. Аналогично $S_2, V_2$ будут зависеть от исходов первых двух бросков монеты. Учитывая 4 возможных исхода, можно записать:
### $V_2(HH) = \Delta_1(H)S_2(HH)+(1+r)(X_1(H)-\Delta_1(H)S_1(H))$
### $V_2(HT) = \Delta_1(H)S_2(HT)+(1+r)(X_1(H)-\Delta_1(H)S_1(H))$
### $V_2(TH) = \Delta_1(T)S_2(TH)+(1+r)(X_1(T)-\Delta_1(T)S_1(T))$
### $V_2(TT) = \Delta_1(T)S_2(TT)+(1+r)(X_1(T)-\Delta_1(T)S_1(T))$
### Есть 6 неизвестных: $V_0,\Delta_0,\Delta_1(H),\Delta_1(T),X_1(H),X_1(T)$. Для $\Delta_1(T)$ можно получить delta-hedging formula:
### $\Delta_1(T)=\frac{V_2(TH)-V_2(TT)}{S_2(TH)-S_2(TT)}$
### Можно решить и для $X_1(T)=\frac{1}{1+r}[\tilde{p}V_2(TH)+\tilde{q}V_2(TT)]$, где вероятности – risk-neutral. Это уравнение – цена опциона в $t=0$, если первая монетка даёт tail = $V_1(T)$.
### $\tilde{p}S_2(TH)+\tilde{q}S_2(TT)=(1+r)S_1(T)$
### Аналогично $X_1(T)=V_1(T)$ и $\Delta_1(T)$ можно сделать для любых комбинаций heads, tail.
### В итоге $\Delta_n, X_n, V_n$ – <font color=7E87FC>stochastic processes</font>, то есть последовательно случайных величин, индексируемых по времени. Индекс снизу будет означать то, от скольки подбрасываний монеты зависит значений этой переменной.
### Стоимость портфеля будет опр. через wealth equation как $X_{n+1}=\Delta_nS_{n+1}+(1+r)(X_n-\Delta_nS_n)$. 
### Случайная переменная $V_2$ будет зависеть от подбрасывания монеты – e.g. $\omega_1\omega_2$, и цена акции будет $S_2(\omega_1\omega_2)$. Тогда для европейского колл-опциона $V_2(\omega_1 \omega_2) = (S_2(\omega_1 \omega_2)-K)^+$.

### Мы хотим опр. $X_0$ и значнеия $\Delta_0, \Delta_1(H), \Delta_1(T)$ так, чтобы $X_2$, полученное рекурсивно, удовлетворяло $X_2(\omega_1 \omega_2)=V_2(\omega_1 \omega_2)$, независимо от значений $\omega_1 \omega_2$.
### <font color=7E87FC>Theorem 1.2.2.</font> (Replication in the multiperiod binominal model):
### Рассмотрим $N$-периодную биноминальную модель ценообразования активов с $0<d<1+r<u$, и $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$. Пусть $V_n$ будет случайной величина (дериватив, делающий payoff в момент $N$), зависищая от первых $N$ бросков монеты ($\omega_1 \omega_2 \ldots \omega_N$):
### $V_n(\omega_1 \omega_2\ldots\omega_n) = \frac{1}{1+r}[\tilde{p}V_{n+1}(\omega_1 \omega_2\ldots\omega_nH)+\tilde{q}V_{n+1}(\omega_1 \omega_2\ldots\omega_nT)]$, где $0 < n < N-1$. Далее определим:
### $\Delta_n(\omega_1 \ldots\omega_n) = \frac{V_{n+1}(\omega_1 \ldots\omega_nH)-V_{n+1}(\omega_1 \ldots\omega_nT)}{S_{n+1}(\omega_1 \ldots\omega_nH)-S_{n+1}(\omega_1 \ldots\omega_nT)}$
### $X_N(\omega_1 \ldots\omega_n)=V_N(\omega_1 \ldots\omega_n) \thinspace \forall \thinspace \omega_1 \ldots\omega_n$
### Мультипериодная биноминальная модель называется <font color=7E87FC>complete</font>, т.к. каждый дериватив может быть реплицирован трейдингом базовым активом и денежным рынком.
### Теорема выше также применяема к <font color=7E87FC>path-dependent options</font>.
### <font color=7E87FC>Пример 1.2.4</font>. Допустим, $S_0=4, u=2, d=\frac{1}{2}, r=\frac{1}{4}, \tilde{p}=\tilde{q}=\frac{1}{2}$. Рассмотрим <font color=7E87FC>lookback</font> (экзотический) опцион, у которого payoff $V_3 = \underset{0≤n≤3}{max}S_n-S_3$ в $t=3$.
### Тогда:
### $V_3(HHH)=S_3(HHH)-S_3(HHH)=32-32=0$
### $V_3(HHT)=S_2(HH)-S_3(HHT)=16-8=8$
### $V_3(HTH)=S_1(H)-S_3(HTH)=8-8=0$
### $V_3(HTT)=S_1(H)-S_3(HTT)=8-2=0$
### $V_3(THH)=S_3(THH)-S_3(THH)=8-8=0$
### $V_3(TTH)=S_2(TH)-S_3(THT)=4-2=2$
### $V_3(TTH)=S_0-S_3(TTH)=4-2=0$
### $V_3(TTT)=S_0-S_3(TTT)=4-9.5=3.5$
### Посчитаем цену опциона в остальные моменты времени, используя backward recursion:
### $V_2(HH)=\frac{4}{5}[\frac{1}{2}V_3(HHH)+\frac{1}{2}V_3(HHT)]=3.2$
### $V_2(HT)=\frac{4}{5}[\frac{1}{2}V_3(HTH)+\frac{1}{2}V_3(HTT)]=2.4$
### $V_2(TH)=\frac{4}{5}[\frac{1}{2}V_3(THH)+\frac{1}{2}V_3(THT)]=0.8$
### $V_2(TT)=\frac{4}{5}[\frac{1}{2}V_3(TTH)+\frac{1}{2}V_3(TTT)]=2.2$
### $V_1(H)=\frac{4}{5}[\frac{1}{2}V_2(HH)+\frac{1}{2}V_2(HT)]=2.24$
### $V_1(T)=\frac{4}{5}[\frac{1}{2}V_2(TH)+\frac{1}{2}V_2(TT)]=1.2$
### $V_0=\frac{4}{5}[\frac{1}{2}V_1(H)+\frac{1}{2}V_1(T)]=1.376$
### Если трейдер продаёт lookback опцион в $t=0$ за $1.376$, то он может хеджировать короткую позицию через покупку $\Delta_0 = \frac{V_1(H)-V_1(T)}{S_1(H)-S_1(T)}=\frac{2.24-1.2}{8-2}=0.1733$ долей акции, что равно $0.6933$ долларам. У трейдера остаётся $1.376-0.6933=0.6827$, которые он может инвестировать на денежном рынке под $25\%$. В $t=1$ у него будет $0.8533$ на денежном рынке. Если акция вырастет до $8$, то акция, которая будет равна $1.3867$, даст ей портфель в $X_1(H)=V_1(H)=2.24$. Если же акция упадёт до $2$, то в $t=1$ её акция будет равна $0.3467$, и портфель $X_1(T)=V_1(T)=1.2$. Продолжая этот процесс, трейдер может быть уверен, что у него будет портфель $V_3$ в $t=3$, вне зависимости от того, как будут падать монеты.

## <font color=7E87FC>1.3 Computational Considerations</font>
### В теореме 1.2.2 сложность вычислений растёт экспоненциально. Т.е. в биноминальной модели со $100$ периодами, будет $2^{100} \approx 10^{30}$ возможных исходов $100$ бросков монеты. Такое просчитывать бред. К счастью, алгоритм из теоремы даёт возможно сделать это более эффективно.
### <font color=7E87FC>Пример 1.3.1</font>. $S_0 = 4, u=2, d=\frac{1}{2}, r=\frac{1}{4}, K=5$ (европейский пут-опцион) с датой экспирации в $t=3$, риск-нейтральные $\tilde{p}=\tilde{q}=\frac{1}{2}$. Payoff будет $V_3=(5-S_3)^+$:
### $V-3(HHH) = 0, V_3(HHT)=V_3(HTH)=V_3(THH)=0$
### $V_3(HTT)=V_3(THT)=V_3(TTH)=3, V_3(TTT)=4.5$
### Всего $2^3=8$ вариантов выше, но есть очевидный simplification. Обозначим как $v_3(s)$ payoff опциона в $t=3$, когда акция равна $s$ в этот момент времени. ## <font color=7E87FC>2.3 Conditional Expectations</font>
### Из биноминальной модели ценообразования вспомним риск-нейтральные вероятности $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$. Получается, что $\frac{\tilde{p}u+\tilde{q}d}{1+r}=1$.
### В любой момент $n$ для всех последовательностей бросков монеты $\omega\ldots\omega_n$ имеем $S_n(\omega_1\ldots\omega_n)=\frac{1}{1+r}[\tilde{p}S_{n+1}(\omega_1\ldots\omega_nH)+\tilde{q}S_{n+1}(\omega_1\ldots\omega_nT)]$, т.е. цена акции в момент $n$ – дисконтированное средневзвешенное двух возможных исходов цены акции в момент $n+1$, где $\tilde{p},\tilde{q}$ – веса для усреднения. В упрощённом виде имеем:
### $\tilde{\mathbb{E}}[S_{n+1}](\omega_1\ldots\omega_n)=\tilde{p}S_{n+1}(\omega_1\ldots\omega_nH)+\tilde{q}S_{n+1}(\omega\ldots\omega_nT)$, или же
### $S_n=\frac{1}{1+r}\mathbb{E}_n[S_{n+1}]$.
### $\tilde{\mathbb{E}}[S_{n+1}]$ – <font color=7E87FC>conditional expectation</font> $S_{n+1}$, основанное на информации в момент $n$ (на знании о первых $n$ исходах броска монетки).
### $\tilde{\mathbb{E}}_1[S_2](H)$ – условное матожидание, а если мы запишем $\tilde{\mathbb{E}}_1[S_2]$ без уточнения интересующего нас исхода броска монетки, у нас будет величина, значение которой будет определно случайным экспериментом.
### В общем виде, если $X$ – случайная величина, зависящая от $N$ первых бросков монетки, мы можем оценить $X$, основываясь на информации, доступные в более ранние $n≤N$ периодов времени.

### Пусть $n$ такое, что $1≤n≤N$, и $\omega_1\ldots\omega_n$ даны. Есть всего $2^{N-n}$ возможных цепочек $\omega_{n+1}\ldots\omega_N$ фиксированной последовательности $\omega_1\ldots\omega_n$. Обозначим за $\#H(\omega_{n+1}\ldots\omega_N)$ – число heads в последовательности исходов и за $\#T(\omega_{n+1}\ldots\omega_N)$ – число tail. Тогда:
### $\tilde{\mathbb{E}}_n[X](\omega_1\ldots\omega_n)=\underset{\omega_{n+1}\ldots\omega_N}{\sum} \tilde{p}^{\#H(\omega_{n+1}\ldots\omega_N)}\tilde{q}^{\#T(\omega_{n+1}\ldots\omega_N)}X(\omega_1\ldots\omega_n\omega_{n+1}\ldots\omega_N)$ и называем эту величину $\tilde{\mathbb{E}}_n[X]$.
### Two extreme cases of conditioning are $\tilde{\mathbb{E}}_0[X]$, the conditional expectation of $X$ based on no information, which we define by $\tilde{\mathbb{E}}_0[X]=\tilde{\mathbb{E}}X$, and $\tilde{\mathbb{E}}_N[X]$, the conditional expectation of $X$ based on knowledge of all $N$ coin tosses, which we define by $\tilde{\mathbb{E}}_N[X]=X$.
### <font color=7E87FC>Теорема 2.3.2</font>. Fundamental properties of conditional expectation: пусть $N$ – положительное целое число, и $X$, $Y$ – случайные величины, зависящие от первых $N$ бросков. Пусть $0≤n≤N$ заданы. Тогда верны следующие свойства:
### 1. <font color=7E87FC>Линейность условных матожиданий</font>. Для всех констант $c_1,c_2$ имеем $E_n[C_1X+c_2Y]=c_1E_n[X]+c_2E_n[Y]$;
### 2. <font color=7E87FC>Выносим известные</font>. Если $X$ действительно только от первых $n$ бросков монеты, тогда $E_n[XY] = X\cdot E_n[Y]$.
### 3. <font color=7E87FC>Iterated conditioning</font>. Если $0≤n≤m≤N$, то $E_n[E_m[X]]=E_n[X]$. В частности, $E[E_m[X]]=Ex$.
### 4. <font color=7E87FC>Независимость</font>. Если $X$ зависит только от $n+1$ бросков через $N$, то $E_n[X]=EX$.
### 5. <font color=7E87FC>Условное неравенство Дженсена</font>. Если $\varphi(x)$ – выпуклая функция фиктивной переменной $x$, то $E_n[\varphi(X)]≥\varphi(E_n[X])$.

## <font color=7E87FC>2.4 Martingales</font>
### Выше у нас было $S_n=\frac{1}{1+r}\tilde{E}_n[S_{n+1}]$.
### Разделим все части на $(1+r)^n$ и получим:
### $\frac{S_n}{(1+r)^n}=\tilde{E}_n\begin{bmatrix}\frac{S_{n+1}}{(1+r)^{n+1}}\end{bmatrix}$. Не важно, напишем ли мы $\frac{1}{(1+r)^{n+1}}$ внутри или снаружи условного матожидания, т.к. по допущения биноминальной теории ставка – константа.
### Учитывая риск-нейтральные меры, для акции без дивидендов, лучшая оценка, основанной на информации в момент $n$, для величины дисконтированной цены акции в момент $n+1$ – дисконтированная цена акции в момент $n$. Такой процесс называется <font color=7E87FC>martingale</font>. Итак, формальное определение...
### Рассмотрим биноминальную модель ценообразования активов. Пусть $M_0,M_1,\ldots, M_N$ – последовательность случайных величин, где каждая $M_n$ зависит от первых $n$ бросков монеты ($M_0 = const$). Такая последовательность случайных переменных – <font color=7E87FC>adapted stochastic process</font>:
### 1. Если $M_n = E_n[M_{n+1}], n=0,1,\ldots,N-1$, то такой процесс – <font color=7E87FC>мартингал</font>;
### 2. Если $M_n≤E_n[M_{n+1}], n=0,1,\ldots,N-1$, то такой процесс – <font color=7E87FC>субмартингал</font>;
### 3. Если же $M_n≥E_n[M_{n+1}], n=0,1,\ldots, N-1$, то такой процесс – <font color=7E87FC>супермартингал</font>.

### Мартингал можте быть не только "one-step-ahead", а также работать для большего количества шагов. Если $M_0,M_1,\ldots,M_N$ – мартингал и $n≤N-2$, то свойства мартингала подразумевают $M_{n+1}=E_{n+1}[M_{n+2}]$. Возьмём условное матожидание с обеих сторон и iterated conditioing property: $E_n[M_{n+1}]=E_n[E_{n+1}[M_{n+2}]] = E_n[M_{n+2}]$
### Тогда $M_n = E_n[M_{n+2}] = E_n[M_m]$ – <font color=7E87FC>multiplestep-ahed</font> версия мартингала.
### Матожидание мартингала постоянно во времени, т.е, если $M_0,M_1,\ldots,M_N$ мартингал, то $M_0=EM_n, n=0,1,\ldots,N$.
### Мартингал имеет тенденцию либо к росту, либо к падению, т.к. матожидание будущих периодов – всегда значение в текущем периоде. А акции растут и падают быстрее, чем денежный рынок, чтобы скомпенсировать риск на non-arbitrage рынке.

### <font color=7E87FC>Теорема 2.4.4</font>. Рассмотрим общий случай биноминальной модели с $0<d<1+r<u$ и риск-нейтральными вероятностями $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$. Тогда, при риск-нейтральных мерах, дисконтированная цена акции – мартингал, т.е. уравнение из начала части про мартингалы выполняется в каждый момент $n$ и для всех последовательностей бросков монеты. То есть $\tilde{\mathbb{E}}_n\begin{bmatrix}\frac{S_{n+1}}{(1+r)^{n+1}}\end{bmatrix}(\omega_1\ldots\omega_n) = \frac{S_n(\omega_1\ldots\omega_n)}{(1+r)^n}$.
### В жизни фондовый рынок имеет больший средний рост, нежели денежный. Инвестор достигает этого засчёт длинных позиций в акциях. На риск-нетрайльных рынках темпы ростов совпадают, т.е. тут сколько не дели и не хеджируй, средний рост будет равен процетной ставке.
### <font color=7E87FC>Теорема 2.4.5</font>. Биноминальная модель с $N$ периодами, $\Delta_0,\Delta_1,\ldots,\Delta_{N-1}$ – adapted процесс портфеля (доля акций), $X_0$ – initial wealth, $X_1,\ldots,X_N$ – wealth process, generated recursively. Тогда дисконтированный процесс капитала $\frac{X_n}{(1+r)^n}, n=0,1,\ldots,N$ – мартингал при риск-нейтральных мерах, т.е. $\frac{X_n}{(1+r)^n} = \tilde{\mathbb{E}}_n\begin{bmatrix}\frac{X_{n+1}}{(1+r)^{n+1}}\end{bmatrix}, n=0,1,\ldots,N-1$.
### Также имеем $\tilde{\mathbb{E}}\frac{X_n}{(1+r)^n}=X_0, n=0,1,\ldots,N$.
### Матожидания мартингала не может меняться со временем и всегда должно быть нулевым (константным) значением мартингала.
### В общем случае, если мы можем найти риск-нейтральную меру (т.е. меру, которая подразумевает, что некоторые price paths имеют нулевую вероятность, и также то, что дисконтированные активы – мартингалы), то нет арбитража в модели – <font color=7E87FC>First Fundamental Theorem of Asset Pricing</font>.
### Рассмотрим $N$-периодную биноминальную модель ценообразования с $0<d<1+r<u$ с безрисковой вероятностной мерой $\tilde{P}$. Пусть $V_N$ – случайная величина (дериватив, делающий payoff в момент $N$), зависящая от броска монеты. Тогда, для $0<n<N$, цена дериватива в момент $n$ опр. риск-нейтральной формулой ценообразования. Более того, дисконтированная цена деривативов – мартингал при мере $\tilde{P}$, т.е. $\frac{V_n}{(1+r)^n}=\tilde{\mathbb{E}}_n\begin{bmatrix} \frac{V_{n+1}}{(1+r)^{n+1}}\end{bmatrix}, n=0,1,\ldots,N-1$.
### Пока что мы обсуждали деривативы, которые дают payoff в одну конкретную дату. Однако есть множество ценных бумаг, такие как купоныне облигации, свопы на процентные ставки, в которых есть series of payments. Для таких имеем следующие формулы ценообразования и хеджирования:
### <font color=7E87FC>Теорема 2.4.8</font>. Cash flow valuation. Рассмотрим $N$-периодную биноминальную модель ценообразования активов с $0<d<1+r<u$ с риск-нейтральной вероятностной мерой $\tilde{P}$. Пусть $C_0,C_1,\ldots,C_N$ – последовательность случайных величин таких, что каждая $C_n$ зависит только от $\omega_1\ldots\omega_n$. Цена в момент времени $n$ дериватива, который также даёт платежи $C_n,\ldots,C_N$, будет:
### $V_n = \tilde{\mathbb{E}}_n\begin{bmatrix} \underset{k=n}{\overset{N}{\sum}} \frac{C_k}{(1+r)^{k-n}}\end{bmatrix}, n=0,1,\ldots,N$.
### $C_n(\omega_1\ldots\omega_n)=V_n(\omega_1\ldots\omega_n)-\frac{1}{1+r}\begin{bmatrix} \tilde{p}V_{n+1}(\omega_1\ldots\omega_nH)+\tilde{q}V_{n+1}(\omega_1\ldots\omega_nT)\end{bmatrix}$
### $\Delta_n(\omega_1\ldots\omega_n)=\frac{V_{n+1}(\omega_1\ldots\omega_nH)-V_{n+1}(\omega_1\ldots\omega_nT)}{S_{n+1}(\omega_1\ldots\omega_nH)-S_{n+1}(\omega_1\ldots\omega_nT)}$, где $0<n<N-1$. Если мы скажем, что $X_0=V_0$, то $X_{n+1}=\Delta_nS_{n+1}+(1+r)(X_n-C_n-\Delta_nS_n)$. Тогда $X_n(\omega_1\ldots\omega_n)=V_n(\omega_1\ldots\omega_n), \thinspace \forall \thinspace n, \omega_n$.

### В теореме 2.4.8 $V_n$ – так называемое "Net Present Value" в момент $n$ последовательности $C_n, \ldots, C_N$. Это сумма значнеий $\tilde{\mathbb{E}}_n\begin{bmatrix} \frac{C_k}{(1+r)^{(n-k)}}\end{bmatrix}$ каждого платежа $C_k$, которые делаются в моменты $k=n,k=n+1,\ldots,k=N$. Важно заметить, что платёж в момент $n$ также включён. $C_n$ зависит от первых $n$ бросков монеты и может быть вынесен из условного матожидания $\tilde{\mathbb{E}}_n$, т.е.
### $V_n=C_n+\tilde{\mathbb{E}}\begin{bmatrix} \underset{k=n+1}{\overset{N}{\sum}} \frac{C_k}{(1+r)^{k-n}}\end{bmatrix}, n=0,1,\ldots,N-1$
### В случае $n=N$ уравнение сокращается до $V_N=C_N$.
### Например, трейдер шортит cash flow $C_0,\ldots,C_n$ (т.е. трейдер доджен делать платежи $C_n$ в каждый момент времени $n$, причём как положительные, так и отрицательные; если платёж $<0$, то трейдер в шорт позиции получает кэш). Допустим, такой трейдер инвестирует в фондовый и денежный рынок так, что в момент $n$, до того как сделать платёж $C_n$, стоимость портфеля $X_n$, и лишь затем он делает платёж $C_n$. Также допустим, что после этого он встаёт в позицию в $\Delta_n$ в акции. Стоимость портфеля в момент $n+1$ до платежа $C_{n+1}$ будет $X_{n+1}$. Если трейдер начинает с $X_0=V_0$ и выбирает позицию в акциях $\Delta_n$, тогда будет работать $X_N=V_N=C_N$. Далее в момент времени $N$ он совершает последний платёж $C_N$ и остаётся с $0$. Он идеально захеджировал шорт позицию в cash flow.

## <font color=7E87FC>2.5 Markov Processes</font>
### Рассмотрим биноминальную модель ценообразования актива. Пусть $X_0,X_1,\ldots,X_N$ – adapted process (the one that cannot see into "the future"). Если для каждого $n\thinspace|\thinspace0<n<N-1$ и для каждой функции $f(x)$ есть такая функция $g(x)$ (зависящая от $n,f$) такая что $E_n[f(X_{n+1})]=g(X_n)$, то $X_0,X_1,\ldots,X_N$ – <font color=7E87FC>Markov process</font>.
### По определению $E_n[f(X_{n+1})]$ – рандомная штука; зависит от первых $n$ бросков монеты. Марковские свойства о том, что эта зависимость от бросков монеты возникает через $X_n$ (т.е. информация о бросках монетки, которые необходимы, чтобы оценить $E_n[f(X_{n+1})]$ is summarized by $X_n$).
### <font color=7E87FC>Пример 2.5.2</font>. Stock price. В биноминальной модели цена акции в момент $n+1$ даётся формулой:
### $S_{n+1}(\omega_1\ldots\omega_n\omega_{n+1})=\begin{cases} uS_n(\omega_1\ldots\omega_n), \text{if } \omega_{n+1}=H, \\ dS_n(\omega_1\ldots\omega_n), \text{if } \omega_{n+1}=T. \end{cases}$
### Тогда $E_n[f(S_{n+1})](\omega_1\ldots\omega_n)=pf(uS_n(\omega_1\ldots\omega_n))+qf(dS_n(\omega_1\ldots\omega_n))$.
### В укороченном виде без омег $E_n[f(S_{n+1})]=g(S_n)$, где функция $g(x)$ фиктивной переменная $x$ опр. как $g(x)=pf(ux)+qf(dx)$. Stock price process is Markov.

# <font color=7E87FC>3. State Prices</font>
## <font color=7E87FC>3.1 Change of Measure</font>
### Рассматриваем конечное выборочное пространство $\Omega$, на котором заданы две вероятностные меры $P$ и $\tilde{P}$. Допустим, обе дают положительную вероятность для каждого элемента в $\Omega$. Тогда можем сделать так: $Z(\omega) = \frac{\tilde{P}(\omega)}{P(\omega)}$
### Поскольку оно зависит от исхода $\omega$ случайного эксперимента, $Z$ – случайная переменная. Это называется <font color=7E87FC>Radon-Nikodym derivative</font> $\tilde{P}$ по $P$, хотя в контексте выборочного пространства это скорее просто quotient (частное), а не производная.
### <font color=7E87FC>Теорема 3.1.1</font>. Пусть $P$ и $\tilde{P}$ – вероятностные меры на конечном выборочном пространстве $\Omega$, допустим также, что $P(\omega)>0$ и $\tilde{P}(\omega)>0$ для всех $\omega \in \Omega$, и определим случайную величину $Z$ при помощи формулы выше. Тогда у этой штуки будут следующие 3 свойства:
### 1. $P(Z>0)=1$;
### 2. $EZ=1$ (пруф: $EZ = \underset{\omega\in\Omega}{\sum}Z(\omega)P(\omega)=\underset{\omega\in\Omega}{\sum}\frac{\tilde{P}(\omega)}{P(\omega)}P(\omega)=\underset{\omega\in\Omega}{\sum}\tilde{P}(\omega)=1$);
### 3. для любой случайной величины $Y$, $\tilde{E}Y=E[ZY]$ (пруф: $\tilde{Y}=\underset{\omega\in\Omega}{\sum}Y(\omega)\tilde{P}(\omega)=\underset{\omega\in\Omega}{\sum}Y(\omega)\frac{\tilde{P}(\omega)}{P(\omega)}P(\omega) = \underset{\omega\in\Omega}{\sum}Y(\omega)Z(\omega)P(\omega)=E[ZY]$).
<img src='https://sun9-56.userapi.com/impg/Xw9aJJVnGrMpp3HXqR6QzRutHUGdzbJwuWN9QQ/LbCqAdZNMmA.jpg?size=1244x702&quality=96&sign=11c0ce3532a40cdfbe6726eeefd43342&type=album' align='right' vpsace='20' hspace='20' width='250'></img>
### Крч в случае с монеткой риск-нейтральные меры $\tilde{p}=\tilde{q}=\frac{1}{2}$, а в случае с обычными, например, $p=\frac{2}{3},q=1-p=\frac{1}{3}$.
### Тогда Radon-Nikodym derivative будет отношением $\tilde{P}$ к $P$.
### $V_0=\tilde{E}\frac{V_3}{(1+r)^3}=E\frac{V_3Z}{(1+r)^3}$ – с использованием случайной величины $Z$. Во втором случае не просто считается матожидание дисконтированного payoff опциона при actual вероятностной мере, но сначала взвешивает эти payoffs с исп. случайной величины $Z$$. Это приводит нас к идее <font color=7E87FC>state price</font>...

### В $N$-периодной биноминальной модели с actual вероятностной мерой $P$ и риск-нейтральными вероятностными мерами $\tilde{P}$, пусть $Z$ – производная Радома-Никодима, т.е. $Z(\omega_1\ldots\omega_N)=\frac{\tilde{P}(\omega_1\ldots\omega_N)}{P(\omega_1\ldots\omega_N)}=(\frac{\tilde{p}}{p})^{\#H(\omega_1\ldots\omega_N)}(\frac{\tilde{q}}{q})^{\#T(\omega_1\ldots\omega_N)}$, где $\#H\ldots$ – число heads в последовательности $\omega_1\ldots\omega_N$ и $\#T\ldots$ – соотв. для tail. Тогда <font color=7E87FC>state price</font> density random variable – это $\xi(\omega)\frac{Z(\omega)}{(1+r)^N}$, и $\xi(\omega)P(\omega)$ – state price, соответсвующая $\omega$.
### Пусть $\bar{\omega}=\bar{\omega}_1\ldots\bar{\omega}_N$ – конкретная последовательность бросков монеты в $N$-периодной модели, и дериватив делает payoff $1$, если $\bar{\omega}$ наступает, иначе payoff $=0$:
### $V_N(\omega)=\begin{cases} 1, \text{if } \omega=\bar{\omega} \\ 0, \text{otherwise}.\end{cases}$
### В соотв. с риск-нейтральной формулой ценообразования, стоимость дериватива в $t=0$ будет $\tilde{E}\frac{V_N}{(1+r)^N}=\frac{\tilde{P}(\bar{\omega})}{(1+r)^N}=\frac{Z(\bar{\omega})P(\bar{\omega})}{(1+r)^N}=\xi (\bar{\omega})P(\bar{\omega})$ – state price в конце = цена в момент ноль дериватива, который платит $1$ в момент $N$ тогда и только тогда, когда $\bar{\omega}$ наступает. Тильда (риск-нейтральность) важна, т.к. без Радона-Никодима и state price, тогда цена дериватива в начале зависела бы только от матожидания дохода и actual probability measure. В более сложном варианте выше цена актива зависит и от матожидания дохода, и от риска ($Z(\bar{\omega})$).
### Однако в жизни большая часть контрактов делает payoffs для нескольких разных значений $\omega$, и payoffs не всегда $=1$. Такой контракт можно рассматривать как портфель из простых контрактов, каждый из которых платит $1$ if and only if наступает $\omega$, и можно просуммировать. In terms of state price:
### $V_0 = E[\xi V_n]=\underset{\omega\in\Omega}{\sum}V_n(\omega)\xi(\omega)P(\omega)$, где $\xi(\omega)$ можно разложить на факторы $\frac{1}{(1+r)^n}$ и $Z(\omega)$.

## <font color=7E87FC>3.2 Radon-Nikodym Derivative Process</font>
### Пусть $Z$ – случайная величина в $N$-периодной биноминальной модели. $Z_n=E_nZ, n=0,1,\ldots,N$. Тогда $Z_n$ – мартингал under $P$.
### Пруф: $E_n[Z_{n+1}]=E_n[E_{n+1}[Z]]=E_n[Z]=Z_n$, вот и мартингал.
### $Z_0=EZ, Z_0=E_0[Z_1]=EZ_1$.
### В $N$-периодной биноминальной модели, пусть определены $P,\tilde{P}$, и предположим, что $P(\omega)>0$ и $\tilde{P}(\omega)>0$ для любой последовательности бросков монеты $\omega$. Также определим производную Радона-Никодима (случайную величину) $Z(\omega)=\frac{\tilde{P}(\omega)}{P(\omega)}$ для любого $\omega$. Тогда процесс Radon-Nikodym derivative будет $Z_n=E_n[Z], n=0,1,\ldots,N$. В частности, $Z_N=Z$ и $Z_0=1$.
### <font color=7E87FC>Lemma 3.2.5</font>. Пусть $n\in\mathbb{Z}|0<n<N$, а $Y$ – случайная величина, зависящая только от первых $n$ бросков монеты. Тогда $\tilde{E}Y=E[Z_nY]$.
### Прфу: $\tilde{E}Y=E[ZY]=E[E_n[ZY]]=E[YE_n[Z]]=E[YZ_n]$.
### $Y(\omega_1\ldots\omega_n\omega_{n+1}\ldots\omega_N)=\begin{cases} 1, \text{if } \omega_1\ldots\omega_n=\bar{\omega_1}\ldots\bar{\omega_n} \\ 0, \text{otherwise.}\end{cases}$
### Тогда $\tilde{E}Y = \tilde{P} \text{ \{the first n coin tosses result in } \bar{\omega_1}\ldots\bar{\omega_n}\} = \tilde{p}^{\#H(\bar\omega_1\ldots\bar{\omega}_n)}\tilde{q}^{\# T(\bar{\omega_1}\ldots\bar{\omega_n})}$.
### Из леммы 3.2.5 получаем $Z_n(\bar{\omega_1}\ldots\bar{\omega_n})=(\frac{\tilde{p}}{p})^{\#H(\bar{\omega_1}\ldots\bar{\omega_n})} (\frac{\tilde{q}}{q})^{\#T(\bar{\omega_1}\ldots\bar{\omega_n})}$
### <font color=7E87FC>Lemma 3.2.6</font>. Пусть $n\in\mathbb{Z}|n≤m, 0<n<N$, и $Y$ – случайная величина, зависящая только от первых $m$ бросков монеты. Тогда $\tilde{E}_n[Y]=\frac{1}{Z_n}E_n[Z_mY]$.

### <font color=7E87FC>Теорема 3.2.7</font>. Рассмотрим $N$-периодную биноминальную модель с $0<d<1+r<u$. Допустим, actual вероятность для heads $=p$, а для tail $=q$, оба больше нуля. Риск-нейтральная вероятности соотв. $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$ также обе больше нуля. Пусть $P,\tilde{P}$ – вероятностные меры. Пусть $Z$ – производная Радона-Никодима $\tilde{P}$ по $P$, а $Z_n, n=0,1,\ldots,N$ – этот процесс.
### Рассмотрим дериватив, дающий payoffs $V_N$, зависящие от всех $N$ бросков монеты. Для $n=0,1,\ldots,N$, цена дериватива в момент $n$ будет $V_n=\tilde{E}_n\frac{V_n}{(1+r)^{N-n}}=\frac{(1+r)^n}{Z_n}E_n\frac{Z_NV_N}{(1+r)^N}=\frac{1}{\xi_n}E_n[\xi_NV_N]$, где state price process опр. как $\xi_n = \frac{Z_n}{(1+r)^n}, n=0,1,\ldots,N$.

## <font color=7E87FC>3.3 Capital Asset Pricing Model</font>
### $CAPM$ базируется на балансе спроса и предложения среди инвесторов, обладающих функцией полезности (utility function), конвертирующей units of consumption в units of satisfaction. Проблема в том, что $CAPM$ исходит из no-arbitrage рынков, и на риск-нейтральном ценообразовании.
### В no-arbitrage pricing, опять же, есть две вероятностные меры – actual и риск-нейтральная. В ценообразовании деривативов мы исп. только риск-нейтральную меру. Actual меру можно исп. для asset management (trade-off между риском и actual ожидаемым доходом) и risk management (вероятность наихудшего исхода).
### Под <font color=7E87FC>utility function</font> будем подразумевать неубывающую (nondecreasing) вогнутую (concave – смотри по второй производной) функцию, определённую на $\mathbb{R}$. Эта функция может принимать значения $[-\infty, +\infty)$. Достаточно часто используют функцию полезности $\ln x$, $x>0$. Функция вогнутая, если $U(\alpha x+(1-\alpha)y)≥\alpha U(x)+(1-\alpha)U(y) \thinspace \forall \thinspace x,y \in \mathbb{R}, \alpha \in (0,1)$.
### $U_p(x)=\begin{cases}\frac{1}{p}(x-c)^p, & \text{if } x>c, \\ 0, & \text{if } 0<p<1 \text{ and } x=c, \\ -\infty, & \text{if } p<0 \text{ and } x=c, \\ -\infty, & \text{if } x<c.\end{cases}$
### Для таких функций <font color=7E87FC>index of absolute risk aversion</font> $-\frac{U''(x)}{U'(x)}$ – гиперболическая функция $\frac{1-p}{x-c}$ при $x>c$. Такой класс функция называем $HARA$ (hyperbolic absolute risk aversion). $HARA$, соответсвующая $p=0$ будет $U_0(x) = \begin{cases}\ln (x-c), & \text{if } x>c \\ -\infty, &\text{if } x≤c. \end{cases}$

### Вогнутость функции нужна для того, чтобы обозначить trade-off между риском и доходностью. Например, игровой автомат даёт $1$ с вероятностью $\frac{1}{2}$ и $99$ с вероятностью $\frac{1}{2}$. Матожидание дохода $50$, но risk-averse агент предпочтёт просто получить $50$, чем играть в рандом. Пусть $X$ – рандомный payoff, т.е. $P(X=1)=P(X=99)=\frac{1}{2}$. Для вогнутой функции полезности у нас работает неравенство Дженсена наоборот (upside down), то есть $EU(X)≤U(EX)$. Действительно, если $U(x)=\ln x$, тогда $E\ln X = 2.30$ и $\ln EX=3.91$.
### Если мы моделируем поведение агента как масимизацию ожидаемой полезности от payoff, то по модели наш агент предпочтёт нерандомный payoff $EX=50$, вместо рандомного payoff $X$. Анализируя принятие решения агентом, мы можем увидеть отношение инвестора к трейд-оффу между риском и доходом.
### Рассмотрим $N$-периодную биноминальную модель с $0<d<1+r<u$. Трейдер начинает с начальным капиталом $X_0$ и хочет инвестировать в фондовый и денежный рынки так, чтобы максимизировать ожидаемую полезность капитала в момент времени $N$. Другими словами, у трейдера есть функция полезности $U$ и хочет решить следующую проблему:
### <font color=7E87FC>Problem 3.3.1</font>. Optimal investment: дан $X_0$, найти adapted portfolio process $\Delta_0, \Delta_1, \ldots, \Delta_{N-1}$, который максимизирует $EU(X_N)$ при wealth equation $X_{n+1}=\Delta_nS_{n+1}+(1+r)(X_n-\Delta_nS_n), n=0,1,\ldots,N-1$. 
### Тут у нас actual вероятностные меры $P$. Агент не склонен к риску (risk-averse) и исп. функцию $U$ для трейд-оффа между действительным риском и действительным доходом. Нет смысла решать это в риск-нейтральных мерах, т.к. в таком случае и фондовый, и денежный рынки имеют одинаковые доходности, и трейдер, максимизирующий $\tilde{E}U(X)$ будет инвестировать только в безрисковый денежный рынок.

### <font color=7E87FC>Пример 3.3.2</font>. $r=\frac{1}{4}$, риск-нейтральная мера: $\tilde{P}(HH)=\tilde{P}(HT)=\tilde{P}(TH)=\tilde{P}(TT)=\frac{1}{4}$.
### Пусть действительная вероятностная мера: $P(HH)=\frac{4}{9}, P(HT)=\frac{2}{9}, P(TH)=\frac{2}{9}, P(TT)=\frac{1}{9}$.
### Рассмотрим трейдера с $X_0=4$ и хочет выбрать $\Delta_0, \Delta_1(H), \Delta_1(T)$, чтобы максимизировать $E\ln X_2$.
### $X_1(H)=8\Delta_0+\frac{5}{4}(4-4\Delta_0)=3\Delta_0+5$
### $X_1(T)=2\Delta_0+\frac{5}{4}(4-4\Delta_0)=-3\Delta_0+5$

### $X_2(HH)=6\Delta_1(H)+\frac{15}{4}\Delta_0+\frac{25}{4}$
### $X_2(HT)=-6\Delta_1(H)+\frac{15}{4}\Delta_0+\frac{25}{4}$
### $X_2(TH)=\frac{3}{2}\Delta_1(T)-\frac{15}{4}\Delta_0+\frac{25}{4}$
### $X_2(TT)=-\frac{3}{2}\Delta_1(T)-\frac{15}{4}\Delta_0+\frac{25}{4}$
### $E\ln X_2=P(HH)\ln X_2(HH) + P(HT)\ln X_2(HT) + P(TH) \ln X_2(TH) + P(TT)\ln X_2(TT)$. Числа и переменные подставляем из вырапжений выше. По сути надо максимизировать это выражение через частные производные.
### $\frac{\delta}{\delta \Delta_0} E\ln X_2 = \frac{15}{4}(\frac{4}{9}\frac{1}{X_2(HH)}+\frac{2}{9}\frac{1}{X_2(HT)}-\frac{2}{9}\frac{1}{X_2(TH)}-\frac{1}{9}\frac{1}{X_2(TT)}) = \frac{5}{12}(\frac{4}{X_2(HH)}+\frac{2}{X_2(HT)}-\frac{2}{X_2(TH)}-\frac{1}{X_2(TT)})$
### $\frac{\delta}{\delta \Delta_1(H)}E \ln X_2 = \frac{4}{9}\frac{6}{X_2(HH)}-\frac{2}{9}\frac{6}{X_2(HT)} = \frac{4}{3}(\frac{2}{X_2(HH)}-\frac{1}{X_2(HT)})$
### $\frac{\delta}{\delta \Delta_1(T)}E \ln X_2 = \frac{3}{2}(\frac{2}{9}\frac{1}{X_2(TH)}-\frac{1}{9}\frac{1}{X_2(TT)})=\frac{1}{6}(\frac{2}{X_2(TH)}-\frac{1}{X_2(TT)})$.
### Приравняв к нулю, получаем следующие три уравнения:
### $\frac{4}{X_2(HH)}+\frac{2}{X_2(HT)}=\frac{2}{X_2(TH)}+\frac{1}{X_2(TT)}$
### $\frac{2}{X_2(HH)}=\frac{1}{X_2(HT)}$
### $\frac{2}{X_2(TH)}=\frac{1}{X_2(TT)}$
### Получаем:
### $X_2(HH)=2X_2(HT)$
### $X_2(TH)=2X_2(TT)$
### В итоге: $X_2(HT)=2X_2(TT)$
### $\Delta_0=\frac{5}{9}, \Delta_1(H)=\frac{25}{54}, \Delta_1(T)=\frac{25}{27}$
### Ожидаемая дисконтированная стоимость портфеля при риск-нейтральной мере всегда равна начальному капиталу: $\tilde{E}\frac{X_N}{(1+r)^N}=X_0$.

### <font color=7E87FC>Problem 3.3.3</font>. Дан $X_0$, найти случайную величину $X_N$ (пофиг на portfolio process), максимизирующую $EU(X_N)$ при $\tilde{E}\frac{X_N}{(1+r)^N}=X_0$.
### $Z(HH)=\frac{\tilde{P}(HH)}{P(HH)}=\frac{9}{16}, Z(HT)=\frac{9}{8},Z(TH)=\frac{9}{8}, Z(TT)=\frac{9}{4}$.
### $\xi_1=\xi(HH)=\frac{Z(HH)}{(1+r)^2}=\frac{9}{25}$
### $\xi_2=\xi(HT)=\frac{18}{25}$
### $\xi_3=\xi(TH)=\frac{18}{25}$
### $\xi_4=\xi(TT)=\frac{36}{25}$

# <font color=7E87FC>Глава 4. American Derivative Securities</font>
## <font color=7E87FC>4.1 Introduction</font>
### Если опцион можно исполнить в момент до даты экспирации, то это <font color=7E87FC>американский опцион</font>. American and European calls have the same price. В случае с опционами пут, так называемая <font color=7E87FC>early exercise premium</font> может быть значительной. Между европейским и американским опционами есть ещё один – <font color=7E87FC>Bermudan option</font>, который разрешает early exercies, но в конкретный список дат.
### Т.к. американский опцион может быть исполнен в любой момент до даты экспирации, он НИКОГДА не будет стоить МЕНЬШЕ, чем payoff по немедленному исполнению – <font color=7E87FC>intrinsic value of the option</font>. В отличие от европейского опциона, у которого процесс дисконтированной цены – мартингал при риск-нейтральной мере, дисконтированная цена американского опциона – супермартингал при той же мере.
### Держатель американского опциона may fail to exercise at the optimal exercise date, и в этом случае опцион имеет свойства терять свою стоимость – свойство супермартингала. В течение любого периода времени, в которой исполнять такой опцион не оптимальон, так или иначе, дисконтированная цена ведёт себя как мартингал.
### По аналогии с европейским, чтобы оценить стоимость американского опциона, нужно условно продать опцион в обмен на начальноый капитал, и дальше думать как использовать этот капитал, чтобы зашортить шорт-позицию по опциону. Наихудший момент для продавца опциона будет наилучшим оптимальным моментом для покупателя опциона, чтобы его исполнить. В итоге получится, что начальная опциона будет равняться стоимости капитала, необходимо, чтобы его захеджировать при оптимальном для покупателя моменте исполнения.

## <font color=7E87FC>4.2 Non-Path-Dependent American Derivatives</font>
### $N$-периодная модель с $0<d<1+r<u$ (non-artitrage condition with $r$). Рассматриваем дериватив с payoff $g(S_N)$ в момент $N$ для некоторой функции $g$. Цена акции – Марковский процесс (эволюция процесса после момента $t$ не зависит от предшествующей $t$ эволюции), запишем стоимость дериватива как $V_n$ в каждый момент $n$ как функцию $v_n$ от цены акции в кокнретный момент, т.е. $V_n=v_n(S_n), n=0,1,\ldots,N$. Риск-нейтральная формула ценообразования и Markov simplification подразумевает, что при $0≤n≤N$, функция $v_n$ опр. как European algorithm:
### $v_N(s)=max\{g(s),0\}$,
### $v_n(s)=\frac{1}{1+r}\begin{bmatrix} \tilde{p}v_{n+1}(us)+\tilde{q}v_{n+1}(ds)\end{bmatrix}, n=N-1,N-2,\ldots,0$, где $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$ – риск-нейтральные вероятности того, что акция пойдёт вверх или вниз соотв. Реплицирующий портфель, хеджирующий шорт-позицию по опциону, будет $\Delta_n=\frac{v_{n+1}(uS_n)-v_{n+1}(dS_n)}{(u-d)S_n}, n=0,1,\ldots,N$.
### Теперь рассмотрим американский дериватив. Payoff функция $g$ дана. В любой момент $n≤N$ держатель дериватива может его исполнить и получить платёж $g(S_n)$ – в этой главе payoff зависит только от текущей цены акции $S_n$ в момент исполнения, а не от траектории цены (price path)).
### Тогда у портфеля, хеджирующего шорт-позицию, всегда должна быть стоимость $X_n$, удовлетворяющая неравенству $X_n≥g(S_n), n=0,1,\ldots,N$.
### Стоимость дериватива в каждый момент $n$ равна КАК МИНИМУМ intrinsic value $g(S_n)$, и стоимость реплицирующего портфеля в этот момент должна быть равна стоимость дериватива. Тут меняет европейский алгоритм на американский:
### $v_N(s)=max\{g(s),0\}$,
### $v_n(s)=max\{g(s),\frac{1}{1+r}\begin{bmatrix} \tilde{p}v_{n+1}(us)+\tilde{q}v_{n+1}(ds)\end{bmatrix}, n=N-1,N-2,\ldots,0$.
### Тогда $V_n=v_n(S_n)$ будет ценой дериватив в момент $n$.

Рассматриваем американский пут опцион с датой экспирации в момент $2$ со страйком $5$. Если первый бросок выдаст heads, то $0.4=v_1(S_1(H))=S_1(H)\Delta_0+(1+r)(X_0-\Delta_0S_0)=8\Delta_0+\frac{5}{4}(1.36-4\Delta_0) = 3\Delta_0+1.7$, т.е. $\Delta_0 = -0.43$.

### <font color=7E87FC>Теорема 4.2.2</font>. Replication of path-independent American derivatives. Рассмотрим $N$-периодную биноминальную модель с $0<d<1+r<u$ с $\tilde{p}=\frac{1+r-d}{u-d}, \tilde{q}=\frac{u-1-r}{u-d}$. Payoff-функция $g(s)$, определим recursively backward in time последовательность функций $v_N(s),v_{N-1}(s),\ldots,v_0(s)$. Также определим $\Delta_n=\frac{v_{n+1}(uS_n)-v_{n+1}(dS_n)}{(u-d)S_n}$, также $C_n=v_n(S_n)=\frac{1}{1+r}[\tilde{p}v_{n+1}(uS_n)+\tilde{q}v_{n+1}(dS_n)]$, где $0<n<N-1$. У нас есть $C_n≥0 \thinspace \forall \thinspace n$. По дефолту будет $X_0=v_0(S_0)$, и тогда стоимости портфеля будут $X_{n+1}=\Delta_nS_{n+1}+(1+r)(X_n-C_n-\Delta_nS_n)$. Тогда у нас будет $X_n(\omega_1\ldots\omega_n)=v_n(S_n(\omega_1\ldots\omega_n)) \thinspace \forall \thinspace n, \omega_1\ldots\omega_n$. В частности, $X_n≥g(S_n) \thinspace \forall \thinspace n$.
## <font color=7E87FC>4.3 Stopping Times</font>
### Естественно, время, когда американский опцион должен быть исполнен, рандомное, т.к. зависит от движений цены базового актива. 
### Опишем "exercise rule" случайной величиной $\tau$:
### $\tau(HH)=\infty, \tau(HT)=2, \tau(TH)=1, \tau(TT)=1$.
<img src='https://sun1-27.userapi.com/impg/IQxaHEL26wKEeUMzRsfprSRNihrgiBKqMQePoQ/jX1bdb8btKo.jpg?size=940x448&quality=96&sign=13039dc8bcd6d5e4f46c815749a7e67a&type=album' vpsace=20 hspace=20 width=300 align=right></img>
### Случайная величина $\tau$ определена на $\Omega = \{ HH,HT,TH,TT\}$ и принимает значения в множестве $\{0,1,2,\infty\}$.
### В $N$-периодной биноминальной модели, <font color=7E87FC>stopping time</font> – случайная величина $\tau$, принимающая значения $0,1,\ldots,N$ или $\infty$ и удовлетворяет условиям, что если $\tau(\omega_1\omega_2\ldots\omega_n\omega_{n+1}\ldots\omega_N)=n$, тогда $\tau(\omega_1\omega_2\ldots\omega_n\omega^{'}_{n+1}\ldots\omega^{'}_N)=n \thinspace \forall \thinspace \omega^{'}_{n+1}\ldots\omega^{'}_N$. Последнее означает, что stopping основывется только на доступной трейдеру информации, т.е. если это происходит в момент $n$, то это основано на первых $n$ бросках монеты.
### Когда у нас есть стохастический процесс и stopping time, тогда мы можем определить <font color=7E87FC>stopped process</font>. Например, пусть $Y_n$ будет процессов дисконтированной цены американского пут опциона, т.е.
### $Y_0=1.36, Y_1(H)=0.32, Y_1(T)=2.4, Y_2(HH)=0, Y_2(HT)=Y_2(TH)=0.64, Y_2(TT)=2.56$.
### Пусть $\tau$ – stopping time. Тогда определим stopped process $Y_{n\land \tau}$ ($\land$ означает минимум из двух данных): $Y_{0\land \tau} = Y_0=1.36$, т.к. $0\land \tau = 0$ вне зависимости от броска монеты, и также с $Y_{1\land \tau}=Y_1$. Тем не менее, $2\land \tau$ зависит от броска монеты; если мы получим $HH$ или $HT$, то $2\land \tau =2$, но если получим $TH$ или $TT$, то будет $2\land \tau = 1$.
### $Y_{2\land \tau}(HH)=Y_2(HH)=0, Y_{2\land\tau}(HT)=Y_2(HT)=0.64, Y_{2\land\tau}(TH)=Y_1(T)=2.4, Y_{2\land\tau}(TT)=Y_1(T)=2.4$
### Важно заметить, что сам процесс не останавливается на $\tau$, но значения именно stopping time замораживаются на $\tau$. Было бы логичнее назвать не stopped, а <font color=7E87FC>frozen process</font>, но уж как есть.
### !!! Процесс дисконтированной цены американского пут опциона – супермартингал, но не мартингал при риск-нейтральной мере. Вторым будет stopped process $Y_{n\land\tau}$.
### <font color=7E87FC>ВАЖНО!</font> Если владелец такого американского опциона пропустил момент, в котором процесс – супермартингал, то момент оптимального исполнения упущен (и дальше идёт мартингал).
### <font color=7E87FC>Теорема 4.3.2</font>. Мартингал, остановившийся на stopping time – мартингал. Супермартинга (суб...), остановившийся на stopping time – супермартингал (суб...).
### <font color=7E87FC>Теорема 4.3.3</font>. Option sampling. Если $X_n$ – субмартингал, то $EX_m≤EX_n$, пока $m≤n$. Это неравенство будет работать, если мы заменим $m$ на $\tau\land n$, где $\tau$ – stopping time. Если супермартингал, то $E_{n\land \tau}≥EX_n$

## <font color=7E87FC>4.4 General American Derivatives</font>
### Тут в общем вводим понятие американского дериватива, чья intrinsic value может быть path-dependent. Всё так же $N$-периодная биноминальная модель с $0<d<1+r<u$. $S_n$ – множество stopping times $\tau$, принимающих значения $\{n,n+1,\ldots,N,\infty\}$. В частности, множество $S_0$ содержит все stopping time. $S_N$ может принимать значение $N$ на некоторых траекториях, также значение $\infty$, и никакие иначе.
### $\forall \thinspace n, n=0,1,\ldots,N$ пусть $G_n$ – случайная величина, зависящая от первых $n$ бросков монеты. Американский дериватив с процессом intrinsic value $G_n$ – контракт, который может быть исполнен в любой момент до (включительно) $N$ и, если исполняется в момент $n$, делает payoff $G_n$. Определим процесс цены $V_n$ для такого контракта формулой риск-нейтрального ценообразования американских деривативов: $V_n=\underset{\tau \in S_n}{max}\tilde{E}_n\begin{bmatrix} I_{\{\tau≤N\}}\frac{1}{(1+r)^{\tau-n}}G_{\tau}\end{bmatrix}, n=0,1,\ldots,N$.
### Идея следующая... допустим, американский дериватив не исполнен в моменты $0,1,\ldots,n-1$, и мы пытаемся оценить его стоимость в момент $n$. В этот момент держатель дериватива может либо исполнить, или же отложить исполнение. Момент $n$ будет зависеть от всех моментов, предшествующим ему, но не после. Другими словами, дата исполнения будет stopping time $\tau$. Так как исполнения не произошло до момента $n$, stopping time будет в $S_n$. Однако, если трейдер не исполняет опцион в принципе ($\tau=\infty$), то он получает $\text{payoff} = 0$. Компонент $I_{\{\tau≤N\}}$ нужен для того, чтобы понять, должен ли $I_{\{\tau≤N\}}\frac{1}{1+r}^{\tau-n}G_{\tau}$ быть заменён на $0$ для тех траекторий, для которых $\tau=\infty$. Когда держатель исполняет в момент $\tau\in S_n$, то стоимость дериватива в момент $n$ будет риск-нейтральным дисконтированным матожиданием payoff'а. Трейдеру нужно выбирать $\tau$, чтобы сделать это как можно большей величиной.
### $V_N=\text{max}\{G_N,0\}$.
### Берём $n=N$ так, что $V_N=\underset{\tau \in S_N}{sup}I_{\{\tau≤N\}}\frac{1}{(1+r)^{\tau-N}}G_{\tau}$.
### Stopping time в $S_N$ принимает значения либо $N$, либо $\infty$, и для такого stopping time будет $I_{\{\tau≤N\}}\frac{1}{(1+r)^{\tau-N}}G_{\tau}=I_{\{\tau=N\}}G_N$.
### Чтобы сделать эту величину как можно большей, нам нужно выбрать $\tau(\omega_1\ldots\omega_N)=N$. есди $G_N(\omega_1\ldots\omega_N)>0$ и $\tau(\ldots)=\infty$, если $G_N(\ldots)≤0$.

### 1. $V_n≥max\{G_n,0\} \thinspace \forall \thinspace n$;
### 2. дисконтированный процесс $\frac{1}{(1+r)^n}V_n$ – супермартингал;
### 3. если $Y_n$ – ещё один процесс, удовл. $Y_n≥max\{G_n,0\} \thinspace \forall \thinspace n$ и для которых $\frac{1}{(1+r)^n}Y_n$ – супермартингал, тогда $Y_n≥V_n \thinspace \forall \thinspace n$.
### Свойство (2) гарантирует то, что трейдер, начинающий с каипталом $V_0$ может собрать хеджирующий портфель со стоимостью $V_n$ в каждый момент $n$. ### <font color=7E87FC>Теорема 4.4.4</font>. Replication of path-dependent American derivatives. Рассмотрим $N$-периодную биноминальную модель ценообразования актива с $0<d<1+r<u$ с $\tilde{p},\tilde{q}$.
### Для каждых $n,n=0,1,\ldots,N$ пусть будет $G_n$ – случайная величина, зависящая от первых $n$ бросков монет. С $V_n, n=0,1,\ldots,N$ можем определить:
### $\Delta_n(\omega_1\ldots\omega_n)=\frac{V_{n+1}(\omega_1\ldots\omega_nH)-V_{n+1}(\omega_1\ldots\omega_nT)}{S_{n+1}(\omega_1\ldots\omega_nH)-S_{n+1}(\omega_1\ldots\omega_nT)}$
### $C_n(\omega_1\ldots\omega_n) = V_n(\omega_1\ldots\omega_n)-\frac{1}{1+r}\begin{bmatrix}  \tilde{p}V_{n+1}(\omega_1\ldots\omega_nH)+\tilde{q}V_{n+1}(\omega_1\ldots\omega_nT) \end{bmatrix}$, где $0≤n≤N-1$. 
### У нас есть $C_n≥0 \thinspace \forall \thinspace n$. $X_0=V_0$ и опр. recursively forward in time стоимости портфеля $X_1,X_2,\ldots,X_N$ как $X_{n+1}=\Delta_nS_{n+1}+(1+r)(X_n-C_n-\Delta_nS_n)$, тогда у нас есть $X_n(\omega_1\ldots\omega_n)=V_n(\omega_1\ldots\omega_n) \thinspace \forall \thinspace n, \omega_1\ldots\omega_n$. В частности, $X_n≥G_n \thinspace \forall \thinspace n$.
### Теорема показывает то, что полученная таким образом цена американского дериватива приемлема для продавца, потому что он собирает хеджирующий портфель для шорт-позиции.
### <font color=7E87FC>Теорема 4.4.5</font>. Optimal exercise. Stopping time $\tau^{*}=min\{n,V_n=G_n\}$ максимизирует одно из уравнений (которое с $I$) выше при $n=0$, т.е. $V_0=\tilde{E}\begin{bmatrix} I_{\{\tau^{*}≤N\}}\frac{1}{(1+r)^{\tau^{*}}} G_{\tau^{*}} \end{bmatrix}$
### Стоимость опциона всегда больше или равна своей intrinsic value. Например, стоимость американского пут опциона всегда $≥0$, но пут всегда будет out of the money (т.е. intrinsic value $<0$ – тогда $\tau=\infty$).

## <font color=7E87FC>5.1 Introduction</font>
### <font color=7E87FC>Симметричное случайное блуждание</font> – дискретная версия Броуновского движения.
<img src='https://sun9-11.userapi.com/impg/x1UiiZTfm9Uaa1ScISN1x8feIWnyzpXi8K3kkg/EaAzcgTlvRw.jpg?size=972x624&quality=96&sign=587b0b5f18791ced118191424752f836&type=album' hspace=20 vspace=20 align=right width=300></img>
### Чтобы построить симметричное случайное блуждание, мы несколько раз подбрасываем fair монетку ($p$ – вероятность $H$ на каждого шагу, и $q=1-p$ – вероятность $T$, все равны $\frac{1}{2}$). Обозначим элементарные исходы как $\omega_1\omega_2\omega_3\ldots$. Пусть
### $X_j = \begin{cases} 1, & \text{if } \omega_j=H, \\ -1,&\text{if } \omega_j=T,\end{cases}$,
### Определим $M_0=0$, $M_n=\underset{j=1}{\overset{n}{\sum}}X_j, n=1,2,\ldots$. Процесс $M_n$ – симметричное случайное блуждание. 
### Если бы вероятности не были бы равны $\frac{1}{2}$, то это было бы <font color=7E87FC>асимметричным</font> случайным блужданием. Симметричный random walk – и мартингал, и Марковский процесс.

## <font color=7E87FC>5.2 First Passage Times</font>
### Симметричное случайное блуждание начинает с $0$ в момент времени $0$. Пусть есть $m\in\mathbb{Z}$ – первый момент времени, когда процесс достигает уровня $m$, т.е. $\tau_m=min\{n;M_n=m\}$. Если процесс никогда не достигает уровня $m$, то $\tau_m=\infty$.
### $\tau_m$ – stopping time, или же <font color=7E87FC>first passage time</font> случайного блуждания до уровня $m$. Определим его распределение. 
### <FONT COLOR=7E87FC>Лемма 5.2.1</FONT>. Пусть $M_n$ – симметричное случайное блуждание. Fix число $\sigma$ и опр. процесс $S_n=e^{\sigma M_n}(\frac{2}{e^{\sigma}+e^{-\sigma}})^n$. Тогда $S_n, n=0,1,2,\ldots$ – мартингал.
### Пусть $n\rightarrow \infty$. Чтобы это сделать, нужно определить предел при $n\rightarrow \infty$ от $e^{\sigma M_n \land \tau_m}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{n\land \tau_m}$. 
### Фиксируем $\sigma>0$ и получаем, что $\underset{n\rightarrow\infty}{lim}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{n\land \tau_m} = \begin{cases} (\frac{2}{e^{\sigma}+e^{-\sigma}})^{\tau_m}, & \text{if } \tau_m < \infty \\ 0, & \text{if } \tau_m=\infty\end{cases}$.
### Правую сторону можно переписать как $I_{\{\tau_m<\infty\}} (\frac{2}{e^{\sigma}+e^{-\sigma}})^{\tau_m}$
### Рассмотрим фактор $e^{\sigma M_n\land \tau_m}$, предположим, что $m>0$ и заметим, что $M_{n\land\tau_m}≤m$, т.к. мы останавливаем мартингал, когда он достигает уровня $m$. Тогда, не важно, конечное или бесконечно $\tau_m$, имеем $0≤e^{\sigma M_n\land \tau_m}≤e^{\sigma m}$.
### Также у нас есть $\underset{n\rightarrow\infty}{lim}e^{\sigma M_n\land \tau_m} = e^{\sigma M_{\tau_m}} = e^{\sigma m}$, если $\tau_m<\infty$.
### Далее $\underset{n\rightarrow\infty}{lim}e^{\sigma M_{n\land\tau_m}}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{n\land\tau_m} = e^{\sigma m}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{\tau_m}$, если $\tau_m<\infty$.
### $\underset{n\rightarrow\infty}{lim}e^{\sigma M_{n\land\tau_m}}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{n\land\tau_m} = I_{\{\tau_m<\infty\}}e^{\sigma m}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{\tau_m}$.
### $E\begin{bmatrix}I_{\{\tau_m<\infty\}}e^{\sigma m}(\frac{2}{e^{\sigma}+e^{-\sigma}})^{\tau_m}\end{bmatrix} = 1$.
### $P\{\tau_m<\infty \}=1$ – симметричное случайное блуждание по-любому достигает уровня $m$.

### <font color=7E87FC>Теорема 5.2.2</font>. Пусть $m$ – abritrary ненулевое целое. Симметричное случайное блуждание достигает уровня $m$ almost surely, т.е. first passage time $\tau_m$ до уровня $m$ is finite almost surely.
### Чтобы определить распределение first passage time $\tau_m$, мы рассматриваем его moment-generating функцию $\varphi_{\tau_m}(u)=Ee^{u\tau_m}$. Для всех $x≥0, e^x=1+x+\frac{1}{2}x^2 + \ldots ≥ x$. Тогда для $u>0$, имеем $e^{u\tau_m}≥u\tau_m$ и $\varphi_{\tau_m}(u)=Ee^{u\tau_m}≥uE\tau_m$. При $u=0$, имеем $\varphi_{\tau_m}(u)=1$. При $u<0$, имеем $\alpha=e^u$ так, что $0<\alpha<1$ и $\varphi_{\tau_m}(u)=E\alpha^{\tau_m}$.
### <font color=7E87FC>Теорема 5.2.3</font>. Пусть $m$ – ненулевое целое. First passage time $\tau_m$ для симметричного случайного блуждания удовл. $E\alpha^{\tau_m}=(\frac{1-\sqrt{1-\alpha^2}}{\alpha})^{|m|} \thinspace \forall \thinspace \alpha\in(0,1)$. Из этой теоремы можем заключить, что $E\tau_m=\infty$.
### <font color=7E87FC>Теорема 5.2.5</font>. Пусть $\tau_1$ – first passage time до уровня $1$ случайного блуждания с вероятностю $p$ для шага вверх и $q=1-p$ для шага вниз. Тогда $P\{\tau_1=2j-1\}=\frac{(2j-2)!}{j!(j-1)!}p^jq^{j-1}, j=1,2,\ldots$.
## <font color=7E87FC>5.3 Reflection Principle</font>
### Рассматриваем симметричное случайное блуждание. Допустим, мы подбрасываем монетку нечётное количество $(2j-1)$ раз. Некоторые траектории случайного блуждания достигнут уровня $1$ за первые $2j-1$ шагов, а некоторые – нет. В случае трёх бросков, есть 8 возможных траекторий, и пять из них достигают уровня $1$. Рассмотрим траекторию, при которой уровень $1$ достигается за $\tau_1≤2j-1$.

Если оригинальная траектория заканчивается выше $1$ в последний момент $2j-1$, то reflected траектория заканчивается ниже $1$, и наоборот. Чтобы посчитать число траекторий, достигающих $1$ к моменту $2j-1$, мы можем посчитать число траекторий, который превышает $1$ в этот момент. ### Для симметричного случайного блуждания имеем $P\{\tau_1≤2j-1\}=P\{M_{2j-1}=1\}+2P\{M_{2j-1}≥3\}$. Также $P\{M_{2j-1}≥3\} = P\{M_{2j-1}≤-3\}$, и $P\{\tau_1≤2j-1\}=P\{M_{2j-1}=1\}+P\{M_{2j-1}≥3\}+P\{M_{2j-1}≤-3\}=1-P\{M_{2j-1}=-1\}$.
### Чтобы $M_{2j-1}$ было $-1$, в первых $2j-1$ бросков должно быть $j-1$ heads и $j$ tails. Есть $\begin{pmatrix} 2j-1 \\ j\end{pmatrix} = \frac{(2j-1)!}{j!(j-1)!}$ тректорий, обладающих такими свойствами, и каждая траектория обладает вероятностью $(\frac{1}{2})^{2j-1}$. 
### $P\{M_{2j-1}=-1\}=(\frac{1}{2})^{2j-1}\frac{(2j-1)!}{j!(j-1)!}$...
## <font color=7E87FC>5.4 Perpetual American Put: An Example</font>
### "Perpetual" (бессрочный) американский опцион пут – тот, у которого нет даты экспирации. Такого инструмента нет в жизни на бирже, это скорее математический концепт, который служит мостом между ценообразованием и хеджированием американских дискретных опционов и их же в контексте непрерывного времени.
### Рассматриваем биноминальную модель с $u=2, d=\frac{1}{2}, r=\frac{1}{4}$. Для такой модель риск-нейтральные вероятностные меры $\tilde{p}=\tilde{q}=\frac{1}{2}$. Цена акции в момент $n$ будет $S_n=S_0\cdot 2^{M_n}$, где $M_n$ – случайное блуждание. При риск-нейтральности, $M_n$ симметрично. Пусть $S_0=4$.

### Рассмотрим американский со страйком $K=4$ и без даты экспирации. В момент $n$ владелец опциона может исполнить его, продав за 4 доллара акцию, которая стоит на рынке $S_n$. Нас, оф корс, интересует стоимость пута как функция от цены базовой акции. Т.к. нет даты экспирации, логично ожидать, что стоимость пута будет зависеть только от $S_n$, а не от времени. То же самое касается optimal exercice policy для покупателя опциона. Рассмотрим несколько вариантов оптимальной стратегии исполнения:
### <font color=7E87FC>Policy 0</font>: сразу же исполняем в stopping time $\tau_0$, в первый раз, когда случайное блуждание $M_n$ достигает уровня $0$, где $\tau_0=0$. Здесь стоимость стратегии исполнения будет $V^{(\tau_0)}=0$.
### <font color=7E87FC>Policy 1</font>: исполняем в первый раз, когда акция падает до уровня $2$. Stopping time будет $\tau_{-1}$, т.е. первый раз, когда случайное блуждание падает до уровня $-1$. Обозначим это как $V^{\tau_{-1}}$.
### <font color=7E87FC>Policy 2</font>: исполняем в первый раз, когда акция падает до уровня $1$. Stopping time будет $\tau_{-2}$ – первый раз, когда случайное блуждание падает до уровня $-2$. Обозначим как $V^{\tau_{-2}}$.

### Пусть $m$ будет положительным целым. Риск-нейтральная стоимость пута, если покупатель использует стратегию исполнения $\tau_{-m}$, когда исполнений происходит в первый раз, когда цена акции падает до $4\cdot 2^{-m}$, тогда $V^{\tau_{-m}}=\tilde{E}\begin{bmatrix} (\frac{1}{1+r})^{\tau_{-m}}(K-S_{\tau_{-m}})\end{bmatrix}=4(1-2^{-m})\tilde{E}\begin{bmatrix}  (\frac{4}{5})^{\tau_{-m}} \end{bmatrix}$.

# <font color=7E87FC>6. Interest-Rate-Dependent Assets</font>
## <font color=7E87FC>6.1 Introduction</font>
### Самый простой fixed income asset – <font color=7E87FC>zero-coupon bond</font> – бонд, который платит номинальную стоимость (<font color=7E87FC>face value</font> или par value) в конкретное время (<font color=7E87FC>maturity</font>). В момент времени до срока погашения, стоимость этого актива меньше, чем номинальная стоимость при условии, что процентная ставка больше нуля. Далее определим доходность (<font color=7E87FC>yeild</font>) бескупонного бонда с данным сроком погашения как постоянную процентную ставку, которая необходима, чтобы цену бонда в момент времени $0$ добить до номинальной стоимости. Т.к. теоретически бескупонный бонд для каждого возможного срока погашения есть теоретическая доходность, соотв. каждому моменту времени, больше чем момент времени $0$. <font color=7E87FC>Yield curve</font> – функция доходности от срока погашения.
### Существование бескупонной облигации в нашей Вселенной, описанной с опционами выше, повышает возможности арбитража.
### В этой главе мы рассматриваем безарбитражный рынок – риск-нейтральные вероятностные меры.

## <font color=7E87FC>6.2 Binominal Model for Interest Rates</font>
### Пусть $\Omega$ – множество $2^N$ возможных исходов $\omega_1\omega_2\ldots\omega_N$ от $N$ бросков монеты, $\tilde{P}$ – вероятностная мера на $\Omega$, при которой каждая последовательность $\omega_1\omega_2\ldots\omega_N$ имеет строго положительную вероятность. Опр. <font color=7E87FC>interest rate process</font> как последовательность случайных величин $R_0,R_1,\ldots,R_{N-1}$, где $R_0$ не явл. случайной, и при $n=1,\ldots,N-1$, $R_N$ зависит только от первых $n$ бросков $\omega_1\ldots\omega_n$. Один доллар, инвестированный в денежный рынок в момент $n$, растёт до $1+R_n$ долларов в момент $n+1$. Хотя процентные ставки случайные, в момент $n$ мы знаем, что процетная ставка будет применяться на денежном рынке в период с $n$ до $n+1$. Это "менее случайная" величина, чем акция, т.к. если мы вложимся в акцию в момент $n$, мы не знаем, сколько она будет стоить в $n+1$.
### Логично предположить, что $R_n>0 \thinspace \forall \thinspace n, \omega_1\ldots\omega_n$. В общем и целом для данной главы важно предположить, что $R_n(\omega_1\ldots\omega_n)>-1$.
### Опр. <font color=7E87FC>discount process</font> как $D_n=\frac{1}{(1+R_0)\ldots(1+R_{n-1})}, n=1,2,\ldots,N, D_0=1$. Важно заметить, что $D_n$ зависит только от первых $n-1$ бросков.
### Риск-нейтральная формула ценообразования говорит нам, что стоимость в момент $0$ платежа $X$, полученного в момент $m$ ($X$ зависит только от $\omega_1\ldots\omega_m$) будет $\tilde{E}[D_mX]$ – риск-нейтральный ожидаемый дисконтированный платёж.
### Исп. эту формулу, чтобы опр. цену в момент $0$ бескупонного бонда, который платит $1$ в срок погашения $m$: $B_{0,m}=\tilde{E}[D_m]$. Доходностью такого бонда будет $y_m$ из $\frac{1}{B_{0,m}}=(1+y_m)^m$. В момент $0$ на один доллар, инвестированный в бонд со сроком погашения $m$, можно будет купить $\frac{1}{B_{0,m}}$ таких бондов. Выведем доходность: $y_m=(\frac{1}{B_{0,m}})^{\frac{1}{m}}-1$.
### <font color=7E87FC>Пример 6.2.1</font>. Допустим, $N=3$, также $\Omega = \{ HHH, HHT, HTH, HTT, THH, THT, TTH, TTT\}$. 
### Далее предположим, что $\tilde{P}\{ HHH \} = \frac{2}{9}, \tilde{P}\{ HHT \} = \frac{1}{9}, \tilde{P}\{ HTH \} = \frac{1}{12}, \tilde{P}\{ HTT \} = \frac{1}{12}, \tilde{P}\{ THH \} = \frac{1}{6}, \tilde{P}\{ THT \} = \frac{1}{12}, \tilde{P}\{ TTH \} = \frac{1}{8}, \tilde{P}\{ TTT \} = \frac{1}{8}$, и сумма этих $\tilde{P}$ равна $1$.
### $A_{HH} = \{\omega_1 = H, \omega_2 = H\} = \{HHH,HHT\}$
### $A_{HT}=\{\omega_1=H, \omega_2=T\} = \{HTH, HTT\}$
### $A_{TH}=\{\omega_1=T, \omega_2=H\} = \{THH, THT\}$
### $A_{TT}=\{\omega_1=T, \omega_2=T\} = \{TTH, TTT\}$
### Также $\tilde{P}\{A_{HH}\}=\frac{1}{3}, \tilde{P}\{A_{HT}\}=\frac{1}{6}, \tilde{P}\{A_{TH}\}=\frac{1}{4}, \tilde{P}\{A_{TT}\}=\frac{1}{4}$.
### И $\tilde{P}\{A_{H}\} = \frac{1}{2}, \tilde{P}\{A_{T}\}=\frac{1}{2}$.

### Пусть $\tilde{P}$ – вероятностная мера на $\Omega$ всех возможных исходов $N$ бросков монет. Допустим, каждая последовательность $\omega_1\ldots\omega_N$ в $\Omega$ имеет положительную вероятность при мере $\tilde{P}$. Пусть $1≤n≤N-1$ и $\bar{\omega_1}\ldots\bar{\omega_N}$. Опр. $\tilde{P}\{\omega_{n+1}=\bar{\omega}_{n+1},\ldots,\omega_N = \bar{\omega}_N|\omega_1=\bar{\omega}_1,\ldots,\omega_n=\bar{\omega}_n\} = \frac{\tilde{P}\{\bar{\omega}_1\ldots\bar{\omega}_n\bar{\omega}_{n+1}\ldots\bar{\omega}_N\}}{\tilde{P}\{\omega_1=\bar{\omega}_1, \ldots, \omega_n=\bar{\omega}_n\}}$.
### Пусть $X$ – случайная величина. Определим условное матожидание $X$, основанное на информации в момент $n$ формулой $\tilde{E}[X](\bar{\omega_1}\ldots\bar{\omega_n})= \underset{\bar{\omega}_{n+1}\ldots\bar{\omega}_n}{\sum} X(\bar{\omega}_1\ldots\bar{\omega}_n\bar{\omega}_{n+1}\ldots\bar{\omega}_N) \times \tilde{P}\{\omega_{n+1}=\bar{\omega}_{n+1},\ldots,\omega_N=\bar{\omega}_N|\omega_1=\bar{\omega}_1, \ldots, \omega_N=\bar{\omega}_n\}$. Далее $\tilde{E}_0[X]=\tilde{E}X$ и $\tilde{E}_N[X]=X$.
### <font color=7E87FC>Zero-coupon bond prices</font>. Пусть $\tilde{P}$ – вероятностная мера на $\Omega$ всех возможных исходов $N$ бросков монет, и допустим, что каждая последовательность $\omega_1\ldots\omega_N$ имеют положительные вероятности при мере $\tilde{P}$. Пусть $R_0,R_1,R_2,\ldots,R_{N-1}$ – процесс процентной ставки, где $R_n$ зависит только от первых $n$ бросков монеты. Опр. процесс дисконтирования $D_n, n=0,1,\ldots,N$. При $0≤n≤m≤N$ цена бескупонного бонда со сроком погашения $m$ в момент $n$ будет $B_{n,m}=\tilde{E}_n\begin{bmatrix} \frac{D_m}{D_n}\end{bmatrix}$.
### $B_{m,m} = 1$ (мы берём номинальную стоимость как $1$ для удобства). В обещм виде при номиналке $C$, стоимость бонда в момент $n$ и сроком погашения $m$ будет $CB_{n,m}$.
### Дисконтированная цена бонда $D_nB_{n,m}, n=0,1,\ldots,m$ – мартингал при мере $\tilde{P}$.
### $\tilde{E}_k[D_nB_{n,m}]=\tilde{E}_k[\tilde{E}_n[D_m]]=\tilde{E}_k[D_m]=D_kB_{k,m}$.
### <FONT COLOR=7E87FC>Теорема 6.2.6</FONT>. Вне зависимости от того, как portfolio random variables $\Delta_{n,m}$ выбираются (subject to the condition that $\Delta_{n,m}$ может зависеть только от первых $n$ бросков монет), процесс дисконтированного капитала $D_nX_n$ – мартингал при мере $\tilde{P}$.

## <font color=7E87FC>6.3 Fixed-Income Derivatives</font>
### Допустим, в биноминальной модели процентных ставок у нас есть актив, цена которого в момент $n$ будет $S_n$. Это может быть акции, но чаще просто контракт, payoff которого зависит от процентной ставки. $S_n$ завиист только от первых $n$ бросков монеты. Риск-нейтральная вероятностная мера $\tilde{P}$, при которой дисконтированная цена актива – мартингал: $D_nS_n=\tilde{E}_n[D_{n+1}S_{n+1}], n=0,1,\ldots,N-1$.
### <font color=7E87FC>Forward conract</font> – обязательство выплатит a specified delivery price $K$ в delivery date $m$, где $0≤m≤N$, для актива, цена которого в момент $m$ равна $S_m$. The $m$-forward price of this asset at time $n$, where $0≤n≤m$, is the value of $K$ that makes the forward contract have no-arbitrage price zero at time $n$.
### <font color=7E87FC>Теорема 6.3.2</font>. Рассмотрим актив с процессом цены $S_N$ в биноминальной модели процентных ставок. Допустим, бескупонная облигация с любым сроком погашения может торговаться на бирже. При $0≤n≤m≤N$, форвардная цена па срок $m$ на момент $n$ этого актива будет $F_{n,m}=\frac{S_n}{B_{n,m}}$.
### Допустим, трейдер в момент $n$ продаёт форвардный контрак с датой исполнения $m$ и ценой исполнения $K$. Цена $K$ выбирается так, чтобы этот форвардный контракт имел цену $0$ в момент $n$. Тогда продажа форвардного контракта не создаёт никакого дохода. После продажи форварда в момент $n$ трейдер сразу же встаёт в шорт-позицию $\frac{S_n}{B_{n,m}}$ бескупонных бондов и использует доход $S_n$, чтобы купить долю актива. Далее трейдер ничего не делает в период до момента $m$, в который он владеет долей актива, которую он поставляет по форвардному контракту. Взамен он получает $K$. После того, как он закрывает короткую позицию, у него остаётся $K-\frac{S_n}{B_{n,m}}$. Если эта величина положительная, то трейдер нашёл арбитраж. Если же отрицательная, то трейдер вместо этого мог сделать всё наоборот, встав в лонг-позицию по форварду, в лонг по $m$-maturity бонду, и зашортить актив, чтобы, опять же, получить арбитраж. Для этого $K$ должна опр. по (6.3.2).
### Это называется <font color=7E87FC>static hedge</font>, т.к. нет активного трейдинга между $n$, когда хедж строится, и момент $m$, когда форвардный контракт исполняется.

### Также можем опр. форвардную процентную ставку. Пусть $0≤n≤m≤N$. Допустим, в момент $n$ трейдер встаёт в короткую позицию по $m$-maturity бонду и исп. доход, полученный от этого, чтобы приобрести $\frac{B_{n,m}}{B_{n,m+1}}$ бескупонных бондов со сроком погашения в $m+1$. Такой портфель строится с нулевыми затратами в момент $n$. В момент $m$ необходимо, чтобы трейдер инвестировал $1$ для покрытия шорт-позиции по $m$-maturity бонду. В момент $m+1$ трейдер получает $\frac{B_{n,m}}{B_{n,m+1}}$ от лонг-позиции по $m+1$-maturity бондам. Тогда в период между $m$ и $m+1$, это тоже самое, как если бы трейдер инвестировал в процентную ставку $F_{n,m}=\frac{B_{n,m}}{B_{n,m+1}}=1=\frac{B_{n,m}-B_{n,m+1}}{B_{n,m+1}}$.
### Заметим, что $F_{m,m} = \frac{B_{m,m}}{B_{m,m+1}}-1=\frac{1}{\frac{1}{1+R_m}}-1=R_m$.
### Процентная ставка $R_m$ может быть залочена в момент $m$ при помощи заимствования в момент $m$.
### <font color=7E87FC>Теорема 6.3.5</font>. Пусть $0≤n≤m≤N-1$. Безарбитражная цена в момент $n$ контракта, который платит $R_m$ в момент $m+1$ будет $B_{n,m+1}F_{n,m}=B_{n,m}-B_{n,m+1}$.
### Допустим, трейдер в момент $n$ продаёт контракт по выплате $R_m$ в момент $m+1$ и получает с этого доход $B_{n,m}-B_{n,m+1}$. Затем он покупает $m$-maturity бонд и встаёт в короткую позицию по $m+1$-maturity бонду. Общая стоимость этого портфеля (шортим один контракт, встаём в лонг по бонду и шортим другой бонд) равна нулю. В момент $m$ трейдер получает доход $1$ по $m$-maturity бонду, и инвестирует эти деньги в денежный рынок под ставку $R_m$. В момент $m+1$, это даёт $1+R_m$. Трейдер платит $1$, чтобы покрыть шорт-позицию по $m+1$-maturity бонду и исп. оставшиеся $R_m$, чтобы сделать pay off по контракту. Получается, что трейдер захеджировал шорт-позицию по контракту, который платит $R_m$ в момент $m+1$. Изначальный капитал, необходимый для этого, будет $B_{n,m}-B_{n,m+1}$ – безарбитражная цена контракта.

### Пусть $m$ будет $1≤m≤N$. $m$-периодный <font color=7E87FC>interest rate swap</font> – контракт, который платит $S_1,\ldots,S_m$ в моменты $1,\ldots,m$ соотв., где $S_n=K-R_{n-1},n=1,\ldots,m$. Фиксированная ставка $K = const$. $m$-периодный своп ставки $SR_m$ – это значение $K$, которое делает безарбитражную у свопа процентной ставки в нулевой момент цену в $0$.
### Трейдер с лонг ("recieve fixed") своп позицией получает постоянные платежи в каждый момент $n$, к такому платежу можно относиться как к фиксированному процентному платежу $K$ на основную сумму в размере $1$, и трейдер производит платежи по переменной процентной ставке $R_{n-1}$ на ту же сумму. Если у трейдера уже есть долг, по которому он совершает платежи по фиксированной процентной ставке, лонг-позиция своп эффективно конвертируется в долг по переменной процентной ставке. Шорт-позицию своп эффективно конвертирует в долг по переменной процентной ставке в долг по фиксированной ставке.
### <font color=7E87FC>Теорема 6.3.7</font>. Безарбитражная цена в нулевой момент на $m$-периодный своп по процентной ставке будет $\text{Swap}_m=\underset{n=1}{\overset{m}{\sum}}B_{0,n}(K-F_{0,n-1}) = K\underset{n=1}{\overset{m}{\sum}}B_{0,n}-(1-B_{0,m})$.
### В частности, $m$-периодный своп ставки будет $S_m=\frac{\sum_{n=1}^m B_{0,n}F_{0,n-1}}{\sum_{n=1}^mB_{0,n}}=\frac{1-B_{0,m}}{\sum_{n=1}^mB_{0,n}}$.
### Пусть дано $m$ такое, что $1≤m≤N$. $m$-периодый <font color=7E87FC>interest rate cap</font> (ограничение процентной ставки)  – контаркт, по которому платятся $C_1,\ldots,C_m$, где $C_n=(R_{n-1}-K)^+, n=1,\ldots,m$. Также есть аналогично <font color=7E87FC>interest rate floor</font>, где $F_m=(K-R_{n-1})^+, n=1,\ldots,m$.
### Контракт, который платит $C_n$ ТОЛЬКО в момент $n$ называется <font color=7E87FC>interest rate caplet</font>, а в случае с $F_n$ – <font color=7E87FC>floorlet</font>.
### Риск-нейтральная цена $m$-периодного кэпа будет $\text{Cap}_m=\tilde{E}\underset{n=1}{\overset{m}{\sum}}D_n(R_{n-1}-K)^+$, а флора – $\text{Floor}_m=\tilde{E}\underset{n=1}{\overset{m}{\sum}}D_n(K-R_{n-1})^+$.
### Заметим, что $K-R_{n-1}+(R_{n-1}-K)^+=(K-R_{n-1})^+$, т.е. в любой момент, payoff порфтеля со свопом и кэпом – то же самое, что payoff флора. В итоге $\text{Swap}_m+\text{Cap}_m=\text{Floor}_m$. То есть при одинаковом $K$ в $m$-периодных свопах ставок, кэп и флор имеют одинаковую изначальную цену.

## <font color=7E87FC>6.4 Forward Measures</font>
### Пусть $V_m$ – payoff в момент $m$ некоторого контракта (либо дериватив, либо базовый актив типа бонда). Риск-нейтральная формула для цены такого контракта в момент $n$, предшествующем $m$, включает условное матожидание $\tilde{E}_n[D_mV_m]$. Риск-нейтральная цена актива в момент $n$ будет $V_n=\frac{1}{D_n}\tilde{E}_n[D_mV_m], n=0,1,\ldots,m$.
### Пусть $m$ = fixed, $1≤m≤N$. Опр. $Z_{m,m}=\frac{D_m}{B_{0,m}}$ и исп. $Z_{m,m}$ для опр. $m$-форвардной меры $\tilde{P}^m$ формулой $\tilde{P}^m(\omega)=Z_{m,m}(\omega)\tilde{P}(\omega) \thinspace \forall \thinspace \omega \in \Omega$.
### $Z_{n,m} = \frac{D_nB_{n,m}}{B_{0,m}}, n=0,\ldots,m$.
### <font color=7E87FC>Теорема 6.4.2</font>. Пусть $m$ = fixed, $1≤m≤N$, $\tilde{P}^m$ – $m$-форвардная мера. Если $V_m$ – случайная переменная, зависящая от первых $m$ бросков, тогда $\tilde{E}^m[V_m]=\frac{1}{B_{0,m}}\tilde{E}[D_mV_m]$. В более общем виде $\tilde{E}_n^m[V_m]=\frac{1}{D_nB_{n,m}}\tilde{E}_n[D_mV_m], n=0,1,\ldots,m$.
### $\tilde{E}_n^m[V_m]=\frac{V_n}{B_{n,m}}, n=0,1,\ldots,m$.
### Последнее – в момент $n$ цена любого дериватива или актива, платящего $V_m$ в момент $m$, denominated in units of the zero-coupon bond со сроком погашения $m$ – форвардная цена актива из (6.3.2).
### В итоге, форвардная цена со сроком исполнения $m$ на недивидендный актив – мартингал при форвардной мере $\tilde{P}^m$.

## <font color=7E87FC>6.5 Futures</font>
### Рассмотрим актив с процессом цены $S_0,S_1,\ldots,S_N$ в биноминальной модели процентных ставок. При $0≤m≤N$, процесс цены фьючерса со сроком исполнения $m$, $\text{Fut}_{n,m},n=0,1,\ldots,m$ – адаптированный процесс со следующими свойствамиЖ
### (1): $\text{Fut}_{m,m}=S_m$
### (2): Для каждого $n, 0≤n≤m≤m-1$, риск-нейтральное значение в момент $n$ по контракту, по которому получают платежи $\text{Fut}_{k+1,m}-\text{Fut}_{k,m}$ в момент $k+1 \thinspace \forall \thinspace k=n,\ldots,m-1$, равно нулю, т.е.:
### $\frac{1}{D_n}\tilde{E}_n\begin{bmatrix} \underset{k=n}{\overset{m-1}{\sum}}  D_{k+1}(\text{Fut}_{k+1,m}-Fut_{k,m})\end{bmatrix} = 0$.
### Трейдер, который в момент $n$ встаёт в лонг-позицию по фьючерсному контракту с доставкой в момент позже, чем $m$, соглашается получать платежи $\text{Fut}_{k+1,m}-F_{k,m}$ в моменты $k+1, k=n,\ldots,m-1$ и также получить актив по его рыночной цене $S_m$ в момент $m$. Платёж может быть отрицательным, в таком случае трейдер в лонг-позиции  должен платить деньги по контракту, чем их получить. Обеспечивается это через margin account, создаваемый брокером для трейдера. В конце каждого дня брокер либо пополняет этот счёт, либо снимает с него деньги. Если баланс счёта становится слишком низким, брокер делает margin call, т.е. просит трейдера либо пополнить счёт, либо закрыть фьючерсные позиции.
### <font color=7E87FC>Теорема 6.5.2</font>. $\text{Fut}_{n,m}=\tilde{E}_n[S_m], n=0,1,\ldots,m$.
